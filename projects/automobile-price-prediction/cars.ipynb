```python
import pandas as pd
```

```python
cars = pd.read_csv('car.csv')
cars
```

```python
cars.columns
```

```python
# Drop columns not used for analysis
cars.drop(['fueltype', 'aspiration', 'enginelocation', 'fuelsystem', 'boreratio', 'enginetype', 'stroke', 'compressionratio', 'peakrpm', 'highwaympg'], axis=1, inplace=True)
```

```python
cars
```

```python
def convertCarBody(data):
    if data == 'convertible':
        return 0
    elif data == 'hardtop':
        return 1
    elif data == 'hatchback':
        return 2
    elif data == 'sedan':
        return 3
    else:
        return 4
```

```python
def convertDriveWheel(data):
    if data == 'fwd':
        return 0
    elif data == 'rwd':
        return 1
    else:
        return 2
```

```python
# 'four', 'six', 'five', 'three', 'twelve', 'two', 'eight'
def convertCylinderNumber(data):
    if data == 'two':
        return 0
    elif data == 'three':
        return 1
    elif data == 'four':
        return 2
    elif data == 'five':
        return 3
    elif data == 'six':
        return 4
    elif data == 'eight':
        return 5
    else:
        return 6
```

```python
def convertDoorNumber(data):
    if data == 'two':
        return 0
    else:
        return 1
```

```python
# Convert Categorical Columns to Numerical
# Doornumber
cars["doornumber"] = cars["doornumber"].apply(convertDoorNumber)
cars
```

```python
# bodytype
cars["carbody"] = cars["carbody"].apply(convertCarBody)
cars
```

```python
# drivewheel
cars["drivewheel"] = cars["drivewheel"].apply(convertDriveWheel)
cars
```

```python
cars["cylindernumber"].unique()
```

```python
cars["cylindernumber"] = cars["cylindernumber"].apply(convertCylinderNumber)
cars
```

```python
# store one for normalization
cars_norm = cars
```

```python
cars.hist()
```

```python
# Normalized data: (x - mean)/range
for i in range(0, len(cars_norm.columns)):
    cars_norm.iloc[:, i] = (cars_norm.iloc[:, i] - cars_norm.iloc[:, i].mean())/(cars_norm.iloc[:, i].max() - cars_norm.iloc[:, i].min())
```
